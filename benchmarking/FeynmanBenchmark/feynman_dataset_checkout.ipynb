```python
import matplotlib.pyplot as plt

import physo.benchmark.FeynmanDataset.FeynmanProblem as Feyn
```

```python
for i in range (1, Feyn.N_EQS+1):
    # Loading problem
    pb = Feyn.FeynmanProblem(i)
    print("\n------------------------ %i : %s ------------------------"%(pb.i_eq, pb.eq_name))
    # Loading data sample
    X_array, y_array = pb.generate_data_points(n_samples = 1000)
    # Plot
    n_dim = X_array.shape[0]
    fig, ax = plt.subplots(n_dim, 1, figsize=(10, 5))
    for i in range(n_dim):
        curr_ax = ax if n_dim == 1 else ax[i]
        curr_ax.plot(X_array[i], y_array, 'k.', )
        curr_ax.set_xlabel("X[%i]" % (i))
        curr_ax.set_ylabel("y")
    plt.show()
    
    print(pb)
```
